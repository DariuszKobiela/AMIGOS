### **Description of AMIGOS  experiments**

1.   **40**  participants watched **16 short** emotional videos while they were **alone**.
2.   The same **40** watched **4 long** videos, **some of them alone** and the **rest in groups**.

**Measures:**
*   Electroencephalogram (EEG)
*   Electrocardiogram (ECG)
*   Galvanic Skin Response (GSR)

**Videos**
*   Frontal,
*   Full-body
*   Depth videos

**Methods for profiling**
*   personality ⇾ Big-five personality traits;
*   mood ⇾ baseline Positive Affect and Negative Affect Schedules;
*   emotions :
    *   **self-assessment** of affective levels (valence, arousal, control, familiarity, like/dislike, and selection of basic emotion) felt by the participants during the **first** experiment,
    *   **external-assessment** of participants' levels of **valence and arousal** for **both** experiments.

**At the beginning** of the recording session of the **(only 1) short** videos experiment, and of each of the two recording sub-sessions of the **long** videos experiment
*   arousal  ------------⇾  “very calm” (1) ---------- (9) “very excited”
*   valence  ------------⇾   “very negative” (1) ---------- (9) “very positive”
*   Dominance         ------⇾   “overwhelmed with emotions” (1) ---------- (9)  “in full control of  emotions”
*   basic emotions   ⇾ Neutral, Disgust, Happiness, Surprise, Anger, Fear, Sadness 


**After film assesment**
*   arousal    -----------⇾ “very calm” (1) ---------- (9) “very excited”
*   valence    -----------⇾“very negative” (1) ---------- (9) “very positive”
*   Dominance   -----⇾“overwhelmed with emotions” (1) ---------- (9)  “in full control of emotions”
*   Like   -----------------⇾ disliking (1) ----------- (9) liking
*   Familiarity      -------⇾“Never seen it before” (1) ----------- (9)   “Know the video very well” 
*   basic emotions   ⇾ Neutral, Disgust, Happiness, Surprise, Anger, Fear, Sadness

**Summary**

 

*   **Short videos – 17** annotations 
  *   16 after each video + 1 at the beginning
*   **Long Videos – 6** annotations 
  *   1 at the beginning + 2 after each two videos + 1 before next two videos + 2 after each video

````
import pandas as pd
import glob
import os

path = r'E:/1_Studia/5_semestr/Advanced_non-relational_databases/PROJECT/data/Divided_Data/'
#print(filenames)

list_of_values_lists = []
counter = 0

emotion_files = ['anger', 'disgust', 'fear', 'happiness', 'neutral', 'sadness', 'surprise']


for emotion_file in emotion_files:
    
    emotion = emotion_file
    filenames = glob.glob(path + emotion_file + '/*.csv')
    
    for filename in filenames:
        counter += 1
        if filename.endswith("ECG.csv"):
            #print('ECG file')
            df_ECG = pd.read_csv(filename)
            ECG_RA_median = df_ECG['ECG_RA'].median()
            ECG_LA_median = df_ECG['ECG_LA'].median()  

        elif filename.endswith("EEG.csv"):
            #print('EEG file')
            df_EEG = pd.read_csv(filename)
            list_EEG = []
            list_EEG.append(df_EEG['AF3'].median())
            list_EEG.append(df_EEG['F7'].median())
            list_EEG.append(df_EEG['F3'].median())
            list_EEG.append(df_EEG['FC5'].median())
            list_EEG.append(df_EEG['T7'].median())
            list_EEG.append(df_EEG['P7'].median())
            list_EEG.append(df_EEG['01'].median())
            list_EEG.append(df_EEG['02'].median())
            list_EEG.append(df_EEG['P8'].median())
            list_EEG.append(df_EEG['T8'].median())
            list_EEG.append(df_EEG['FC6'].median())
            list_EEG.append(df_EEG['F4'].median())
            list_EEG.append(df_EEG['F8'].median())
            list_EEG.append(df_EEG['AF4'].median())
            df_EEG_cumulated = pd.DataFrame(list_EEG, columns=['EEG medians'])
            EEG_value = df_EEG_cumulated['EEG medians'].mean()

        else:
            #print('GSR file')
            df_GSR = pd.read_csv(filename)
            GSR_median = df_GSR['GSR_RAW'].median()

        if counter%3 == 0:
            list_of_values_lists.append([ECG_RA_median, ECG_LA_median, GSR_median, EEG_value, emotion])
        
final_data = pd.DataFrame(list_of_values_lists, columns=['ECG_RA', 'ECG_LA', 'GSR', 'EEG', 'emotion'])
````

**•	Naive Bayes Classifier** - It is a classification technique based on Bayes’ Theorem with the assumption of independence among predictors. In other words, Naive Bayes classifiers assume that the presence of a particular feature in a class is unrelated to the presence of any other feature or that all of these properties have independent contribution to the probability.

**•Decision Trees** -  A decision node has two or more branches and a leaf node represents a classification or decision. The topmost decision node in a tree which corresponds to the best predictor called root node. Decision trees can handle both categorical and numerical data.

**•	Nearest Neighbor** - To label a new point, it looks at the labelled points closest to that new point (those are its nearest neighbors). Closeness is typically expressed in terms of a dissimilarity function. Once it checks with ‘k’ number of nearest neighbors, it assigns a label based on whichever label the most of the neighbors have.

**•	Support Vector Machines** - SVM training algorithm builds a model that assigns new examples to one category or the other, making it a non-probabilistic binary linear classifier. An SVM model is a representation of the examples as points in space, mapped so that the examples of the separate categories are divided by a clear gap that is as wide as possible. New examples are then mapped into that same space and predicted to belong to a category based on the side of the gap on which they fall.

**Data Preparation & Data Preprocessing**

We loaded the data prepared in the previous sprints, then slightly reformatted it (removing the unnecessary column and got rid of the empty values), then we  indexed the rows correctly.

We also swapped labels with emotions like "anger, neutral etc." on numeric labels.

````
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r'./data/final_data.csv', index_col=None)
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
df.reset_index(level=None, drop=True, inplace=True)

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder() 
df['emotion'] = encoder.fit_transform(df['emotion']) # fit_transform
X = df.iloc[:,:-1]
y = df.emotion

X.shape, y.shape
````

image.png

**Division of the set & Model training**

We divided the data into a training set (80%) and a test set (20%).

Together we selected the parameters of the tree and trained it.

````
from sklearn.model_selection import train_test_split

seed = 999

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)

X_train.shape, X_test.shape, y_train.shape, y_test.shape

from sklearn.metrics import recall_score, \
                            precision_score, \
                            accuracy_score, \
                            confusion_matrix, \
                            classification_report, \
                            plot_roc_curve

def classification_scores(trues, preds):
    return print('Accuracy {}\nRecall: {}\nPrecision: {}\n'.format(accuracy_score(trues, preds),
                                                                  recall_score(trues, preds, average='weighted'), 
                                                                  precision_score(trues, preds, average='weighted')))
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier(max_depth=4, random_state=seed) #wywoluje model

tree_model.fit(X_train, y_train) #trenuje model

tree_pred = tree_model.predict(X_test) #testuje model

classification_scores(y_test, tree_pred)
````

We have made an attempt to refine the parameters.

````
from sklearn.model_selection import RandomizedSearchCV

params = {'max_depth': [3,4,5,6],
         'criterion': ['gini', 'entropy'],
         'max_features': [3,6,10,12],
         'min_samples_split': [10,20]}

tree_rand_search = RandomizedSearchCV(tree_model, 
                                     param_distributions=params,
                                     n_iter=100,
                                     n_jobs=-1,
                                     random_state=seed)

tree_rand_search.fit(X_train, y_train)
tree_rand_search.best_score_
````

3) We have also standardized data through built-in function to have smaller values.

4) We evaluated **bagging algorithm** for classification - we came up with this idea after consulting the results of the first experiment with the subgroup that dealt with the implementation of the sprint at KNIME. Unfortunately, the introduction of bagging did not improve our result.

````
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'./data/final_data_skipped.csv', index_col=None)
df.head(10)
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
df.reset_index(level=None, drop=True, inplace=True)
df_ready = df.drop(['ECG_LA'], axis=1)
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder() 
df['emotion'] = encoder.fit_transform(df['emotion']) 
X = df.iloc[:,:-1]
y = df.emotion

X.shape, y.shape

from sklearn.model_selection import train_test_split

seed = 999

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)

X_train.shape, X_test.shape, y_train.shape, y_test.shape


from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
ct = ColumnTransformer([('standardize', StandardScaler(), ['ECG_RA', 'GSR', 'EEG'])], remainder='passthrough')
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
````

**Metrics**

````
from sklearn.metrics import recall_score, \
                            precision_score, \
                            accuracy_score, \
                            confusion_matrix, \
                            classification_report, \
                            plot_roc_curve
def classification_scores(trues, preds):
    return print('Accuracy {}\nRecall: {}\nPrecision: {}\n'.format(accuracy_score(trues, preds),
                                                                  recall_score(trues, preds, average='weighted'), 
                                                                  precision_score(trues, preds, average='weighted')))
# trenujemy drzewo, na razie wstawiamy parametry jakie nam się podoba
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier(max_depth=4, random_state=seed) #wywoluje model

tree_model.fit(X_train, y_train) #trenuje model

tree_pred = tree_model.predict(X_test) #testujemy model
classification_scores(y_test, tree_pred)
````

##### **Bagging Algorithm**

````
#BAGGING
# evaluate bagging algorithm for classification
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import BaggingClassifier
# define the model
bagging_model = BaggingClassifier()
# FIT
bagging_model.fit(X_train, y_train)
# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

n_scores = cross_val_score(bagging_model, X_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
````

**•	RNN** is the most proper learning model for learning sequential input data and the time-series data classification where the feedback and the present value is fed again into the network and the output contains the adding of values in the memory. At each time step, the RNN receives an input, updates its hidden state, and makes a prediction. RNN uses a gradient descent algorithm through time for training the weights. 

**•	LSTM** (long short term memory) - is a specific type of traditional RNN designed for temporal sequences of data and the long-range dependencies which depend on each other e.g. text, songs, speech etc. LSTM uses memory blocks instead of simple RNN units where each memory block includes one or more memory cells with a pair of adaptive multiplicative gates as the input and output. A memory block places information and updates them across time-steps based on the input and output gates. The gates control the input and output flow of information to a memory cell. In simple words the current value of a sequence depends on a value which was in an earlier sequence (like in our EKG case).

**•	CNN** (convolutional neural networks) - CNN is a network trained by a gradient-based optimization algorithm used for detection - usually with photos and movies to detect some features of them. It can also be used with visualizations like a sonogram of speech or EKG to detect emotions.

## Time Series Neural Network Model

While creating an emotion-recognition algorithm using RNN, we encountered a lot of problems that we couldn't overcome. Finally, we prepared the data for training neural networks (training set, validation set, test set), but each of the algorithms used refused to obey.

We did some of the LSTM tutorials, but in this time of the sprint and without substantive help, we were unable to make the code work. 

We have found many examples like this below but all of them was not proper to our main goal - multiclassification.

````
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import CuDNNLSTM, Dense, Dropout, LSTM
from keras.optimizers import Adam
(X_train, y_train),(X_test, y_test) = mnist.load_data()  # unpacks images to x_train/x_test and labels to y_train/y_test
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
classifier = Sequential()
#Adding the input LSTM network layer
classifier.add(CuDNNLSTM(128, input_shape=(X_train.shape[1:]), return_sequences=True))
classifier.add(Dropout(0.2))
#Adding a second LSTM network layer
classifier.add(CuDNNLSTM(128))
#Adding a dense hidden layer
classifier.add(Dense(64, activation='relu'))
classifier.add(Dropout(0.2))
#Adding the output layer
classifier.add(Dense(10, activation='softmax'))
classifier.add(LSTM(128, input_shape=(X_train.shape[1:]), return_sequences=True))
#Compiling the network
classifier.compile( loss='sparse_categorical_crossentropy',
              optimizer=Adam(lr=0.001, decay=1e-6),
              metrics=['accuracy'] )
#Fitting the data to the model
classifier.fit(X_train,
         y_train,
          epochs=3,
          validation_data=(X_test, y_test))
test_loss, test_acc = classifier.evaluate(X_test, y_test)
print('Test Loss: {}'.format(test_loss))
print('Test Accuracy: {}'.format(test_acc))
````

# **Conclusions and recommendations**

On the last project meeting we have been discussing the lessons that we have learned during this 3-month period. This project made us realize how important it is to do your research about a problem that you are facing.

However 2 most important lessons or advice for future researchers would be:

##### 1. In your further research, you can try to shuffle the data and repeat our experiments (code is available in our github repository). It may slightly improve the results.

After transforming the data from Matlab format to CSV we have decided that it would be a great idea to sort these files into emotion directories. At first it seemed logical, the more organized set the better classification, because we know where it came from. However after some experiments we realized that this type of organization imposes that we first take into consideration in our model videos with detected anger then fear etc. Next team of researchers should firstly take care of the shuffling the data set so that the models had a better opportunity of learning.

##### 2. You should look into the problem differently. As an output, we can have multiple emotions (1, 2, 3 ...). This is a very hard problem called multiclassification. 

We think that we may have been too ambitious about the subject, considering that this is our first project of this type. We rushed into deep water and tried to create a classifier that allows us to recognize between as many as 7 emotions! This is a lot, so if we were to start our project from scratch, learned from experience and poor results, we would start with a simple binary classifier that would allow us to recognize between two emotions: positive and negative. Maybe it is too simple for someone experienced, but step by step we, as a begginers, would try to explore the topic and strive for multiclassification and a positive impulse in the form of a good binary classifier would be something that would give us a desire for further action.
